```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
collision = pd.read_csv("Motor_Vehicle_Collisions_Crashes.csv")
```

```python
collision.head()
```

```python
collision['datetime'] = pd.to_datetime(collision['CRASH DATE'] + ' ' + collision['CRASH TIME'])

```

```python
collision['Number_of_injuries'] = collision['NUMBER OF PERSONS INJURED'] + collision['NUMBER OF PEDESTRIANS INJURED'] + collision['NUMBER OF CYCLIST INJURED'] + collision['NUMBER OF MOTORIST INJURED']
collision['Number_of_death'] = collision['NUMBER OF PERSONS KILLED'] + collision['NUMBER OF PEDESTRIANS KILLED'] + collision['NUMBER OF CYCLIST KILLED'] + collision['NUMBER OF MOTORIST KILLED']
collision['Total_injury_death'] = collision['NUMBER OF PERSONS INJURED'] + collision['NUMBER OF PEDESTRIANS INJURED'] + collision['NUMBER OF CYCLIST INJURED'] + collision['NUMBER OF MOTORIST INJURED'] + collision['NUMBER OF PERSONS KILLED'] + collision['NUMBER OF PEDESTRIANS KILLED'] + collision['NUMBER OF CYCLIST KILLED'] + collision['NUMBER OF MOTORIST KILLED']

```

```python
collision.drop(['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'], axis=1, inplace = True)

```

```python
collision.dropna(subset=['BOROUGH'],inplace=True)
collision.dropna(subset=['ZIP CODE'],inplace=True)
```

```python
collision['Year'] = collision.datetime.dt.year
collision['Month'] = collision.datetime.dt.month
collision['Day'] = collision.datetime.dt.day
collision['Hour'] = collision.datetime.dt.hour
collision['Day_of_week'] = collision.datetime.dt.dayofweek

```

```python
collision
```

```python
collision = collision[collision['ZIP CODE'] != '     ']
```

```python
collision[['ZIP CODE']] = collision[['ZIP CODE']].astype(int)
```

```python
collision.to_csv("collision_data_clean.csv")
```
